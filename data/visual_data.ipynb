```python
import pandas as pd
import random
import os
import dill
```

```python
data_path = "./processed_data/pretrain_lab_data.pkl"
data_df = pd.read_pickle(data_path)
```

```python
def add_name(row):
    typ = row["TYPE"]
    if typ == "MED":
        item_id = row["NDC"]
    elif typ in ["DIAG", "PROC"]:
        item_id = row["ICD9_CODE"]
    elif typ == "LAB":
        item_id = row["ITEMID"]
    idx = vocab.get_word_id(item_id, typ)
    row["NAME"] = vocab.idx2detail[idx]
    return row
```

```python
subject_id = data_df["SUBJECT_ID"][random.randint(0, len(data_df))]
data_df = data_df[data_df["SUBJECT_ID"] == subject_id]
```

```python
type_len = 3
types = list(data_df["TYPE"].unique())
type_dfs = []
for t in types:
    t_df = data_df[data_df["TYPE"] == t]
    start_id = random.randint(0, len(t_df) - type_len)
    type_dfs.append(t_df[start_id: start_id + type_len])
show_df = pd.concat(type_dfs)
show_df = show_df.apply(add_name, axis=1)
```
